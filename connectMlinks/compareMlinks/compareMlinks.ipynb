```python
sdf = silodf[['mlink','Hour','MATSIM volumes']]
sdf.columns = ['mlink','hour','volume']
```

```python
tem = np.array(sdf[(sdf['mlink'] == 76) & (sdf['hour'] == 12)].volume)
```

```python
np.sort(tem)
```

```python
resdict = {
    'mlink':[],
    'hour':[],
    'vol':[]
}
mlinklist = np.unique(sdf.mlink)
for ml in tqdm_notebook(mlinklist):
    for hour in range(1,25):
        vollist = np.array(sdf[(sdf['mlink'] == ml) & (sdf['hour'] == hour)].volume)
        vollist  = np.sort(vollist)
        volen = len(vollist)
        tvol = 0
        if volen >=3:
            tvol = vollist[volen//2]
        else:
            tvol = vollist[-1]
        resdict['mlink'].append(ml)
        resdict['hour'].append(hour)
        resdict['vol'].append(tvol)
```

```python
gsdf = pd.DataFrame(resdict)
```

```python
gsdf
```

```python
resdf = gsdf.merge(avgmdf, on=["mlink","hour"])
```

```python
resdf
```

```python
resdf = resdf[['mlink','hour','vol','volume']]
resdf.columns = ['mlink','hour','matvol','vol']
resdf = resdf[(resdf['matvol']!=0)& (resdf['vol']!=0)]
```

```python
resdf['proportion'] = resdf['vol'] / resdf['matvol']
```
