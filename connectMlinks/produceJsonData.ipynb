```python
import json
import pandas as pd
from tqdm import tqdm_notebook
```

```python
df = pd.read_csv('./data/full-link-mlink-count-primary.csv')
links = pd.read_csv('./data/linknode.csv')
```

```python
links = links.set_index('id')
```

```python
links
```

```python
df
```

```python
res = {}
for idx in tqdm_notebook(df.index):
    lid = df.at[idx, 'Link Id']
    key = str(lid)
    if key not in res:
        desc = ['linkid:' + str(lid), 'mlink:' + str(df.at[idx, 'Link Id']),'Route:'+ df.at[idx, 'Route_Name'], str(df.at[idx, 'Hour']) + ":" +str(df.at[idx, 'MATSIM volumes'])]
        desc = '<br/>'.join(desc)
        res[key] = {
            'desc':desc,
            'loc': [links.at[lid, 'y0'],links.at[lid, 'x0']]
        }
    else:
        res[key]['desc'] += "<br/>" + str(df.at[idx, 'Hour']) + ":" +str(df.at[idx, 'MATSIM volumes'])
```

```python
jsonstr = json.dumps(res)
jsonstr = "var mydata = " + jsonstr + ";"
with open('./data/data.js','w') as f:
    f.write(jsonstr)
```
