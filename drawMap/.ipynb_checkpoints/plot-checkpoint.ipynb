```python
# import shapefile as shp
import matplotlib.pyplot as plt
import geopandas as gp
import pandas as pd
# sf = shp.Reader("PerthSA1.shp")
```

```python
wa = gp.read_file("PerthSA1.shp")
```

```python
dt = pd.read_csv('./comp2031.csv')
```

```python
wa
```

```python
wa.index = wa['PerthSA1_S'].astype(int)
```

```python
dt.drop(['pop1','area2','pop2'])
```

```python
for idx, row in dt.iterrows():
    area = row['area1']
    val = row['diff']
    wa.at[area, 'diff']= val
```

```python
wa
```

```python
fig, ax = plt.subplots(1, 1,figsize=(15,15),dpi=300)

wa.plot(column='diff', ax=ax, legend=True, cmap='RdBu')
```
