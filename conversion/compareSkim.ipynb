```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mga50codes = pd.read_csv('./data/mga50codes.csvwithHeaders.csv', index_col  = 0)
```

```python
google = pd.read_csv('./data/SA1TravelTime3.csv')
```

```python
google['duration'] = google['duration']/60
```

```python
google['durationskim'] = -1.0
```

```python
for idx,row in google.iterrows():
    ori = int(row['origin'])
    dest = int(row['destination'])
    tt = mga50codes.at[ori,str(dest)]
    google.at[idx, 'durationskim'] = tt
```

```python
google
```

```python
plt.figure(figsize=(10,10),dpi=200)
plt.plot(google['duration'], google['durationskim'], 'ro',markersize=1)
plt.axis([0, 40, 0, 40])
plt.xlabel('Google travel time')
plt.ylabel('Matsim travel time')
plt.show()
```
